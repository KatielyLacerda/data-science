```python
#Katiely Feitosa de Lacerda, número USP: 12777100
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# URL dos dados
url = "https://www.worldometers.info/world-population/population-by-country/"
#primeira questão, ler os dados tabela, calcular a matriz de correlação e ver as mais negatigas
# Ler os dados da tabela HTML
dados = pd.read_html(url)[0]
```

```python
# calcula a matriz de correlação das variáveis da tabeça
corr = dados.corr()

#Plot Correlation Matrix using Matplotlib
plt.figure(figsize=(7, 5))
plt.imshow(corr, cmap='Blues', interpolation='none', aspect='auto')
plt.colorbar()
plt.xticks(range(len(corr)), corr.columns, rotation='vertical')
plt.yticks(range(len(corr)), corr.columns);
plt.suptitle('Correlation between variables', fontsize=15, fontweight='bold')
plt.grid(False)
plt.show()
```

```python
#lendo o arquivo
df = pd.read_csv("Life Expectancy Data.csv") #reading the file
```

```python
#conta os nulos
print(df.isnull().sum())
```

```python
varnames = df.columns
```

```python
#definindo os paises
countries = np.unique(df['Country'])
countries
```

```python
#criando e plotando as séries temporais dos países nos anos
selected_countries = ['Ireland', 'Finland', 'Netherlands', 'Zambia', 'Angola']
filtered_data = df[df["Country"].isin(selected_countries)]
variable = varnames[3]
plt.figure(figsize=(12, 6))
for c in selected_countries:
    country_data = filtered_data[filtered_data["Country"] == c]
    life_expectancy = country_data[variable]
    date = country_data['Year']
    plt.plot(date,life_expectancy, label = c)

plt.title("Time series")
plt.xlabel("Year")
plt.ylabel(variable)
plt.legend()
plt.show()
```

```python
#lendo o arquivo
df = pd.read_csv("Life_Expectancy_Data.csv")
```

```python
#Selecione os dados da coluna 'Life_expectancy'
dados_life_expectancy = df['Life_expectancy ']

# Crie o histograma com os dados da culuna
plt.figure(figsize=(6, 4))
num_bins = 10
n, bins, patches = plt.hist(dados_life_expectancy, num_bins, facecolor='blue', alpha=0.5, density=True, edgecolor='black', linewidth=1.2)
plt.xlabel("Life_expectancy ", fontsize=15)
plt.ylabel("Frequência", fontsize=15)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

```python
#calcula a média de Life_expectancy
media = df["Life_expectancy "].mean()
print(f"A média da Life_expectancy é: {media:.2f}")  # Exibindo com 2 casas decimais

#calcula a variância de Life_expectancy
variancia = df["Life_expectancy "].var()
print(f"A variância da Life_expectancy é: {variancia:.2f}")  # Exibindo com 2 casas decimais
```
