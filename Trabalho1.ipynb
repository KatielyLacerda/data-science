```python
import random
random.seed(1)
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
import pandas as pd  # Adicione esta linha

#lendo os dados com erro de novo
data = pd.read_csv('iris-with-errors.csv', header=(0))
print("NÃºmero de linhas e colunas:", data.shape)
data.head(25)
```

```python
#agora onde tem NaN ou ? vamos subistituir pela mÃ©dia de cada atributo p. Primeiro vamos apagar os ? e colocar o NaN no lugar
import numpy as np
data = data.replace('?', np.nan)
data.head(25)
```

```python

# convertemos para o formato Numpy e ignoramos a Ãºltima coluna
Data = np.array(data[data.columns[0:data.shape[1]-1]], dtype = float)
# calculamos a mÃ©dia de cada atributo e armazenamos em um vetor
# Usamos a funÃ§Ã£o nanmean que calcula a mÃ©dia ignorando os Nan
averages = np.nanmean(Data, axis = 0)
for i in np.arange(0, Data.shape[0]):
    for j in np.arange(0, Data.shape[1]):
        if(np.isnan(Data[i,j]) == True): # verificar se Ã© uma entrada com Nan
            Data[i,j] = averages[j] # insere a mÃ©dia
print(Data)
```

```python
#agora vamos calcular a mediana de cada coluna
medians = data.median()
print(medians)
```

```python
import random
random.seed(1)
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
import pandas as pd  # Adicione esta linha

# CSV file
data = pd.read_csv('train.csv', header=(0))
print("NÃºmero de linhas e colunas:", data.shape)
```

```python
corr = data.corr()
# Mostra a matriz
plt.figure(figsize=(7, 7))
# mostra as imagens
plt.imshow(corr, cmap='Blues', interpolation='none', aspect='auto')
# barra lateral de cores
plt.colorbar()
# inclui o nome das variÃ¡veis
plt.xticks(range(len(corr)), corr.columns, rotation='vertical')
plt.yticks(range(len(corr)), corr.columns);
plt.suptitle('CorrelaÃ§Ã£o entre variÃ¡veis', fontsize=15, fontweight='bold')
plt.grid(False)
plt.show()
```

```python
#fala quais sÃ£o as vÃ¡riaveis mais correlacionadas usando o parÃ¢metro p como correlaÃ§Ã£o mÃ­nima
p = 0.25 # correlaÃ§Ã£o mÃ­nima
var = []
for i in corr.columns:
    for j in corr.columns:
        if(i != j):
            if corr[i][j] > p: # se maior do que p
                var.append([i,j])
print('VariÃ¡veis mais correlacionadas:\n', var)
```
