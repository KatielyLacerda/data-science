```python
import random
random.seed(1)
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
import pandas as pd  # Adicione esta linha

#lendo os dados com erro de novo
data = pd.read_csv('iris-with-errors.csv', header=(0))
print("Número de linhas e colunas:", data.shape)
data.head(25)
```

```python
#agora onde tem NaN ou ? vamos subistituir pela média de cada atributo p. Primeiro vamos apagar os ? e colocar o NaN no lugar
import numpy as np
data = data.replace('?', np.nan)
data.head(25)
```

```python

# convertemos para o formato Numpy e ignoramos a última coluna
Data = np.array(data[data.columns[0:data.shape[1]-1]], dtype = float)
# calculamos a média de cada atributo e armazenamos em um vetor
# Usamos a função nanmean que calcula a média ignorando os Nan
averages = np.nanmean(Data, axis = 0)
for i in np.arange(0, Data.shape[0]):
    for j in np.arange(0, Data.shape[1]):
        if(np.isnan(Data[i,j]) == True): # verificar se é uma entrada com Nan
            Data[i,j] = averages[j] # insere a média
print(Data)
```

```python
#agora vamos calcular a mediana de cada coluna
medians = data.median()
print(medians)
```

```python
import random
random.seed(1)
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
import pandas as pd  # Adicione esta linha

# CSV file
data = pd.read_csv('train.csv', header=(0))
print("Número de linhas e colunas:", data.shape)
```

```python
corr = data.corr()
# Mostra a matriz
plt.figure(figsize=(7, 7))
# mostra as imagens
plt.imshow(corr, cmap='Blues', interpolation='none', aspect='auto')
# barra lateral de cores
plt.colorbar()
# inclui o nome das variáveis
plt.xticks(range(len(corr)), corr.columns, rotation='vertical')
plt.yticks(range(len(corr)), corr.columns);
plt.suptitle('Correlação entre variáveis', fontsize=15, fontweight='bold')
plt.grid(False)
plt.show()
```

```python
#fala quais são as váriaveis mais correlacionadas usando o parâmetro p como correlação mínima
p = 0.25 # correlação mínima
var = []
for i in corr.columns:
    for j in corr.columns:
        if(i != j):
            if corr[i][j] > p: # se maior do que p
                var.append([i,j])
print('Variáveis mais correlacionadas:\n', var)
```
